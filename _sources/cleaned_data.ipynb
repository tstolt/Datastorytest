```python
df3 = pd.read_csv("Vehicle_and_Boat_Registrations_by_Fuel_Type_per_County.csv")
df3.drop(df3[df3['Record Type'] == 'BOAT'].index, inplace = True)
df3.drop(df3[df3['Record Type'] == 'TRL'].index, inplace = True)
nyc_counties = ['NEW YORK', 'KINGS', 'BRONX', 'RICHMOND', 'QUEENS']
df3 = df3[df3['County'].isin(nyc_counties)]

fuel_other = ['COMP N/G', 'FLEX', 'OTHER', 'NONE', 'PROPANE']
fuel_dieselgas = ['DIESEL', 'GAS']
df3['Fuel Type'] = df3['Fuel Type'].replace(fuel_other, 'OTHER')
df3['Fuel Type'] = df3['Fuel Type'].replace(fuel_dieselgas, 'GAS/DIESEL')

df3
```

```python
df5 = pd.read_excel('epa_pm25.xlsx') 
agg = df5.groupby('SCC LEVEL 1')['Emissions (Tons)'].sum()
aggregated_data_modified = agg.to_frame()
categories_to_sum = {
    'Highway Vehicles': ['Highway Vehicles - Compressed Natural Gas (CNG)',
                         'Highway Vehicles - Diesel',
                         'Highway Vehicles - Electricity',
                         'Highway Vehicles - Ethanol (E-85)',
                         'Highway Vehicles - Gasoline'],
    'Off-highway Vehicles': ['Off-highway Vehicle CNG',
                             'Off-highway Vehicle Diesel',
                             'Off-highway Vehicle Gasoline',
                             'Off-highway Vehicle LPG'],
    'Other': ['Agriculture Production - Crops',
             'Agriculture Production - Livestock',
             'Open Burning',
             'Pleasure Craft',
             'Mining and Quarrying: SIC 10 and SIC 14',
             'Railroad Equipment',
             'Other Combustion',
             'Marine Vessels, Commercial']}
for category, columns in categories_to_sum.items():
    category_sum = aggregated_data_modified.loc[columns].sum()
    aggregated_data_modified = aggregated_data_modified.drop(columns)
    aggregated_data_modified.loc[category] = category_sum
aggregated_data_modified
```
